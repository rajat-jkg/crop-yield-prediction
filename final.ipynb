```python
import pandas as pd
import matplotlib.pyplot as mlp
from sklearn import linear_model , svm
from datetime import datetime
```

## Wheat Model

```python
pd.options.mode.copy_on_write = True
dfWheat = pd.read_csv('wheat-data.csv')[11:]
wheatModel = linear_model.LinearRegression()
wheatModel.fit(dfWheat[['Rainfall_jan_april']], dfWheat.hg_per_ha_yield)
```

## Maize Model

```python
dfMaize = pd.read_csv('maize-data.csv')[11:]
maizeModel = svm.SVR()
maizeModel.fit(dfMaize[['Rainfall_june_aug']], dfMaize.Yeild_hg_per_ha)
```

## Paddy Model

```python
dfPaddy = pd.read_csv('Rice-Data.csv')[11:]
paddyModel = svm.SVR()
paddyModel.fit( dfPaddy[['Rainfall_june_to_december']], dfPaddy.hg_per_ha_yield)
```

### Prediction Script

```python
## Rainfall Data Input
wheatRain = int(input("Rainfall from Jan to Apr"))
maizeRain = int(input("Rainfall from June to Aug"))
riceRain = int(input("Rainfall form June to Dec"))
```

## Random Predictions 

```python
print("Predicted wheat yeild",round(wheatModel.predict([[wheatRain]])[0],2))
print("Predicted maize yeild",round(maizeModel.predict([[maizeRain]])[0],2))
print("Predicted paddy crop yeild",round(paddyModel.predict([[riceRain]])[0],2))
```
